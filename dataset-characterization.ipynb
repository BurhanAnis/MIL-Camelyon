```python
import numpy as np
import matplotlib.pyplot as plt
import openslide
from packages.utils import get_list_of_paths
```

```python
slide_paths_pos = get_list_of_paths('/Volumes/BurhanAnisExtDrive/camelyon/camelyon_data/training/positive/images')
slide_paths_neg = get_list_of_paths('/Volumes/BurhanAnisExtDrive/camelyon/camelyon_data/training/negative')
slide_paths_test = get_list_of_paths('/Volumes/BurhanAnisExtDrive/camelyon/camelyon_data/test/images')
```

```python
#count dimensions
def get_all_dims(slide_paths):
    dims = []
    for path in slide_paths:
        slide = openslide.OpenSlide(path)
        width, height = slide.level_dimensions[0]
        dims.append((height, width))
    return dims 
```

```python
training_dims = get_all_dims(slide_paths_pos + slide_paths_neg)
```

```python
# Extract height and width values for each dataset
heights, widths = zip(*training_dims)

plt.figure(figsize=(20, 12))
plt.scatter(widths, heights, color = 'Green', alpha = 0.3)

plt.ylabel("Height /px")
plt.xlabel("Width /px")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(widths, bins = 8, color = 'green', alpha = 0.5)
plt.xlabel('Widths in pixels')
plt.ylabel('Frequency')
```

```python
plt.hist(heights, bins = 8, color = 'green', alpha = 0.5)
plt.xlabel('Heights in pixels')
plt.ylabel('Frequency')
```
